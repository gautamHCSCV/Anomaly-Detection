```python
import numpy as np
import os
import cv2
import pyod
```

```python
from deepface import deepface
from tensorflow import keras
```

```python
base_model = deepface.create_deepface()
weights_path = deepface.get_weights()
base_model.load_weights(weights_path)
```

```python
X_train = []
```

```python
path = "/content/classfication final data/No Defect/"
images = os.listdir(path)
X_train = list(X_train)
print(len(X_train))
for i in range(300,500):
    img1 = cv2.imread(path+images[i])
    img1 = img1.astype('float32')
    img1 = cv2.resize(img1, dsize=(152, 152), interpolation=cv2.INTER_AREA)
    img1 = img1.reshape(1, 152, 152, 3)
    X_train.append(base_model(img1)[0])
X_train = np.array(X_train)
X_train.shape
```

```python
from pyod.models.ecod import ECOD
clf = ECOD()
clf.fit(X_train)
y_train_scores = clf.decision_scores_
```

```python
print(np.mean(y_train_scores))
y_train_scores[50:100]
```

```python
X_test = []
```

```python
path1 = "/content/classfication final data/Foreign/"
test_images = os.listdir(path1)
X_test = list(X_test)
for i in range(25,50):
    img1 = cv2.imread(path1+test_images[i])
    img1 = img1.astype('float32')
    img1 = cv2.resize(img1, dsize=(152, 152), interpolation=cv2.INTER_AREA)
    img1 = img1.reshape(1, 152, 152, 3)
    X_test.append(base_model(img1)[0])
X_test = np.array(X_test)
X_test.shape
```

```python
y_test_scores = clf.decision_function(X_test)
print(np.mean(y_test_scores))
y_test_scores
```

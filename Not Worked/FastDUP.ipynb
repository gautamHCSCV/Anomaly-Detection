```python
import os
import pandas as pd

def create_dataframe_from_images(root_dir):
    image_paths = []
    labels = []

    for label in os.listdir(root_dir):
        label_dir = os.path.join(root_dir, label)

        if os.path.isdir(label_dir):
            for image_file in os.listdir(label_dir):
                image_path = os.path.join(label_dir, image_file)
                image_paths.append(image_path)
                labels.append(label)

    data = {'filename': image_paths, 'label': labels}
    df = pd.DataFrame(data)
    return df

# Specify the root directory of your image dataset
root_directory = '/content/classfication final data/'

# Create the dataframe
df = create_dataframe_from_images(root_directory)
df.head()
```

```python
fd = fastdup.create(input_dir="/content/classfication final data/",
                    work_dir="/content/drive/MyDrive/Classification/report/")
fd.run(annotations=df, overwrite=True)
```

```python
fd.vis.similarity_gallery(slice='diff' ,num_images = 35, threshold = 0.95)
```

```python
fd.vis.duplicates_gallery()    # create a visual gallery of duplicates
fd.vis.outliers_gallery()      # create a visual gallery of anomalies
fd.vis.component_gallery()     # create a visualization of connected components
fd.vis.stats_gallery()         # create a visualization of images statistics (e.g. blur)
```
